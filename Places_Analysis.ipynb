```python
import pandas as pd
```

```python
Places_Analysis = pd.read_csv("PlacesAnalysis.csv")
Places_Analysis
```

```python
# virgülü noktaya çevirip sayıya dönüştürdüm
Places_Analysis['User Rating (1-5)'] = Places_Analysis['User Rating (1-5)'].str.replace(',', '.').astype(float)
Places_Analysis.head()
```

```python
%pip install matplotlip
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Görselleştirmeler için stil ayarı
sns.set(style="whitegrid")

# 1. Fiyat Dağılımı
price_distribution = Places_Analysis['Price Range Unit'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=price_distribution.index, y=price_distribution.values, palette="viridis")
plt.title("Fiyat Aralığına Göre Mekan Sayısı")
plt.xlabel("Fiyat Aralığı (Unit)")
plt.ylabel("Mekan Sayısı")
plt.show()

# 2. Puan Dağılımı
plt.figure(figsize=(8, 5))
sns.histplot(Places_Analysis['User Rating (1-5)'], bins=10, kde=True, color="blue")
plt.title("Kullanıcı Puanlarının Dağılımı")
plt.xlabel("Puan (1-5)")
plt.ylabel("Frekans")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=Places_Analysis['User Rating (1-5)'], color="orange")
plt.title("Kullanıcı Puanlarının Boxplot Analizi")
plt.xlabel("Puan (1-5)")
plt.show()

# 3. Etkinlik Türü
event_distribution = Places_Analysis['Event Type'].value_counts()

plt.figure(figsize=(8, 5))
plt.pie(event_distribution.values, labels=event_distribution.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.title("Etkinlik Türüne Göre Mekan Dağılımı")
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```
